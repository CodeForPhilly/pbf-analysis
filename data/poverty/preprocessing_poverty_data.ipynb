```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data_poverty/ACSST5Y2018.S1701_data_with_overlays_2020-11-02T143849.csv", header = 0)
```

```python
df.head()
```

```python
df['total_population'] = df['S1701_C01_001E']
df['below_poverty'] = df['S1701_C02_001E']
df['percent_below_poverty'] = df['S1701_C03_001E']
df['unrelated_individuals_population'] = df['S1701_C01_046E']
df['unrelated_individuals_below_poverty'] = df['S1701_C02_046E']
df['unrelated_individuals_percent_below_poverty'] = df['S1701_C03_046E']
df['income_below_50percent_poverty_level'] = df['S1701_C01_038E']
df['income_below_125percent_poverty_level'] = df['S1701_C01_039E']
df['income_below_150percent_poverty_level'] = df['S1701_C01_040E']
df['income_below_185percent_poverty_level'] = df['S1701_C01_041E']
df['income_below_200percent_poverty_level'] = df['S1701_C01_042E']
df['income_below_300percent_poverty_level'] = df['S1701_C01_043E']
df['income_below_400percent_poverty_level'] = df['S1701_C01_044E']
df['income_below_500percent_poverty_level'] = df['S1701_C01_045E']
df['mean_income_deficit_unrelated_individuals'] = df['S1701_C01_058E']
```

```python
# double check the columns
columns = ['total_population', 
           'below_poverty', 
           'percent_below_poverty', 
           'unrelated_individuals_population',
           'unrelated_individuals_below_poverty',
           'unrelated_individuals_percent_below_poverty',
           'income_below_50percent_poverty_level',
           'income_below_125percent_poverty_level',
           'income_below_150percent_poverty_level',
           'income_below_185percent_poverty_level',
           'income_below_200percent_poverty_level',
           'income_below_300percent_poverty_level',
           'income_below_400percent_poverty_level',
           'income_below_500percent_poverty_level',
           'mean_income_deficit_unrelated_individuals']
for col in columns:
    print(col)
    print(df.at[0,col], '\n')
```

```python
# drop PA state-level information 
df.drop([0,1], inplace = True)

# get cleaned zipcode
df["zipcode"] = df["GEO_ID"].apply(lambda x: x[-5:])
df.rename(columns = {"GEO_ID": "zipcode"})

columns.insert(0, "zipcode")
df = df[columns]

df.reset_index(drop = True, inplace = True)
```
